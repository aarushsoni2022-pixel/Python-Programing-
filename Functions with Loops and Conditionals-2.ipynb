```python
#11.Write a function to check if a number is an Armstrong number.
def arm(n):
    temp =n
    total=0
    digits=len(str(n))

    while temp>0:
        digits = temp%10
        total +=digits**digits
        temp //=10
    if total==n:
        return "armstrong number"
    else:
        return "not armstrong number"
num=int(input("enter a number:"))
print(arm(num))
    
```

```python
#12.Write a function to generate the Fibonacci series up to n terms.
def feb(n):
    a,b=0,1
    for i in range(n):
        print(a,end=" ")
        a,b=b,a+b
num=int(input("enter the number of terms"))
print(feb(num))
```

```python
#13.Write a function to check if a number is a palindrome.
def palin(n):
    if str(n) == str(n)[::-1]:
        return "palindrome"
    else:
        return "not palindrome"
num=int(input("enter a number"))
print(palin(num))
```

```python
#14.Write a function to count the number of vowels in a string.
def vol(n):
    count=0
    for ch in n:
        if ch.lower() in "aeiou":
            count +=1
    return count
num=input("enter a string ")
print(vol(num))
```

```python
#15.Write a function to find the sum of digits of a number.
def sum(n):
    total=0
    for digit in str(n):
        total +=int(digit)
    return total 
num=int(input("enter the number"))
print(sum(num))
```
