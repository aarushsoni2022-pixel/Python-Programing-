```python
#31.Write a function to count uppercase and lowercase letters in a string.
def count_letter_cases(input_string):
    
    
    uppercase_count = 0
    lowercase_count = 0
    
    for char in input_string:
        if char.isupper():
            uppercase_count += 1
        elif char.islower():
            lowercase_count += 1
    
    return uppercase_count, lowercase_count

# Example usage
text = "Hello World! Python is AWESOME 123."
upper, lower = count_letter_cases(text)
print(f"Uppercase letters: {upper}")
print(f"Lowercase letters: {lower}")

```

```python
#32.Write a function to check if a string is a palindrome.
text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#33.Write a function to remove all punctuation from a string.
text = input("Enter a string: ")

result = ""
for ch in text:
    if ch.isalnum() or ch.isspace():
        result += ch

print(result)
```

```python
#35.Write a function to replace vowels in a string with a specific character.
text = input("Enter a string: ")
char = input("Enter a character to replace vowels: ")

result = ""
for ch in text:
    if ch.lower() in "aeiou":
        result += char
    else:
        result += ch

print("Modified string:", result)
```

```python
#36.Write a function to find the most frequent character in a string.
text = input("Enter a string: ")

freq = {}
for ch in text:
    if ch != " ":
        freq[ch] = freq.get(ch, 0) + 1

most_frequent = max(freq, key=freq.get)
print("Most frequent character:", most_frequent)

```

```python
#38.Write a function to check if two strings are anagrams.
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

if sorted(str1.replace(" ", "").lower()) == sorted(str2.replace(" ", "").lower()):
    print("Anagrams")
else:
    print("Not Anagrams")
```

```python
#39.Write a function to remove spaces from a string.
def remove_spaces(text):
    return text.replace(" ", "")

# Example
s = "Hello World Python"
print(remove_spaces(s))
```

```python
#40.Write a function to find the longest word in a sentence.
def longest_word(sentence):
    words = sentence.split()
    longest = ""

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest

# Example
s = "Python is a powerful programming language"
print(longest_word(s))
```
