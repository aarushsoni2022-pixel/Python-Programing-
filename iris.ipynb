```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris.isnull().sum()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris["species"].value_counts()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.boxplot(x="species", y="sepal_length", data=iris, ax=axes[0,0])
axes[0,0].set_title("Sepal Length by Species")

sns.boxplot(x="species", y="sepal_width", data=iris, ax=axes[0,1])
axes[0,1].set_title("Sepal Width by Species")

sns.boxplot(x="species", y="petal_length", data=iris, ax=axes[1,0])
axes[1,0].set_title("Petal Length by Species")

sns.boxplot(x="species", y="petal_width", data=iris, ax=axes[1,1])
axes[1,1].set_title("Petal Width by Species")

plt.tight_layout()
plt.show()
```

```python
iris.hist(figsize=(10,8))
plt.show()
```

```python
sns.pairplot(data=iris,hue="species",diag_kind="kde")
plt.show()
```

```python
sns.boxplot(data=iris,x="species",y="sepal_length")
plt.show()
```

```python
sns.violinplot(data=iris,x="species",y="petal_length")
plt.show()
```

```python
sns.lineplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
plt.show()
```

```python
sns.boxenplot(data=iris,x="species",y="petal_width")
plt.show()
```

```python
iris_setosa=iris[iris["species"]=="setosa"]
sns.histplot(data=iris_setosa,x="petal_length",bins=20,kde=True)
plt.title("Setosa")
plt.show()
```

```python
sns.displot(iris, x="sepal_length", bins=10, kde=True, col="species",height=3.7,aspect=0.8)
plt.show()
```

```python
g = sns.FacetGrid(iris, col="species")
g.map_dataframe(sns.kdeplot, x="sepal_width",color="r")
plt.show()
```

```python
iris_mean=iris.groupby("species").agg({"petal_length":["mean","max","min","std"]})
iris_mean.plot(kind="line",figsize=(8,6))
```

```python
# Scatterplot: Sepal length vs Sepal width
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.lmplot(data=iris,x="sepal_width",y="petal_width",hue="species")
plt.show()
```

```python
sns.regplot(data=iris,x="sepal_length",y="petal_length")
plt.show()
```
