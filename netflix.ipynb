```python
plt.figure(figsize=(12,8))

sns.countplot(
    data=df,
    x="release_year",
    hue="type",
    palette="Set2"
)

plt.title("Number of Movies and TV Shows Released Each Year", fontsize=14)
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.xticks(rotation=90)


plt.show()
```

```python
sns.countplot(data=df, x="type")
plt.title("Count of Movies vs TV Shows")
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="rating", order=df['rating'].value_counts().index)
plt.xticks(rotation=40)
plt.title("Content Count by Rating")
plt.show()
```

```python
sns.histplot(df["release_year"], bins=20, kde=True)
plt.title("Distribution of Release Years")
plt.show()
```

```python
top_countries = df["country"].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Content Producing Countries")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
sns.boxplot(data=df, x="type", y="release_year")
plt.title("Release Year Distribution by Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(
    data=df,
    x="type",
    y="release_year"
)
plt.title("Release Year Distribution by Type")
plt.show()

```

```python
df['year_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.year

plt.figure(figsize=(12,5))
sns.countplot(data=df, x="year_added")
plt.xticks(rotation=90)
plt.title("Content Added to Netflix Each Year")
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.show()

```
