```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("tips.csv")
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="time",
    size="size",
    sizes=(20, 200),
    alpha=0.7
)

plt.title("Total Bill vs Tip (by Time & Group Size)")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="day",
    y="tip",
    hue="sex"
)

plt.title("Tip Distribution by Day and Gender")
plt.show()
```

```python
df["tip_percent"] = (df["tip"] / df["total_bill"]) * 100

plt.figure(figsize=(12,5))

sns.barplot(
    data=df,
    x="day",
    y="tip_percent",
    estimator="mean",
    ci=None
)

plt.title("Average Tip Percentage by Day")
plt.ylabel("Tip %")
plt.show()
```

```python
sns.pairplot(
    df,
    hue="sex",
    diag_kind="kde"
)
plt.show()
```

```python
g = sns.FacetGrid(df, col="day", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.pointplot(
    data=df,
    x="day",
    y="tip",
    hue="sex",
    ci="sd"
)

plt.title("Average Tip Trend by Day and Gender")
plt.show()
```

```python
pivot = df.pivot_table(
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)

sns.heatmap(pivot, annot=True, cmap="coolwarm")
plt.title("Average Tip Heatmap (Day vs Time)")
plt.show()
```
