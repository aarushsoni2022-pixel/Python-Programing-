```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = pd.read_csv("Titanic-Dataset.csv")
```

```python
sns.countplot(data=df, x="Survived")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(data=df, x="Survived", hue="Sex")
plt.title("Survival by Gender")
plt.show()
```

```python
sns.barplot(data=df, x="Pclass", y="Survived", estimator="mean")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
sns.boxplot(data=df, x="Survived", y="Age")
plt.title("Age Distribution by Survival")
plt.show()

```

```python
sns.violinplot(data=df, x="Sex", y="Age", hue="Survived", split=True)
plt.title("Age Distribution by Gender & Survival")
plt.show()
```

```python
sns.histplot(df["Age"].dropna(), kde=True, bins=30)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="Age",
    y="Fare",
    hue="Survived",
    alpha=0.6
)
plt.title("Age vs Fare (Survival)")
plt.show()
```

```python
sns.pointplot(
    data=df,
    x="Pclass",
    y="Survived",
    hue="Sex"
)
plt.title("Survival Trend by Class and Gender")
plt.show()
```

```python
g = sns.FacetGrid(df, col="Pclass", hue="Survived")
g.map(sns.scatterplot, "Age", "Fare")
g.add_legend()
plt.show()
```
