```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = pd.read_csv("WineQT.csv")
```

```python
sns.countplot(data=df, x="quality")
plt.title("Distribution of Wine Quality")
plt.show()
```

```python
sns.boxplot(data=df, x="quality", y="alcohol")
plt.title("Alcohol Content by Wine Quality")
plt.show()
```

```python
sns.violinplot(data=df, x="quality", y="pH")
plt.title("pH Distribution by Wine Quality")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="alcohol",
    y="density",
    hue="quality",
    palette="viridis"
)
plt.title("Alcohol vs Density (by Quality)")
plt.show()
```

```python
sns.regplot(
    data=df,
    x="alcohol",
    y="quality",
    scatter_kws={"alpha":0.5}
)
plt.title("Regression: Alcohol vs Quality")
plt.show()
```

```python
sns.pairplot(
    df,
    vars=["alcohol", "pH", "density", "fixed acidity"],
    hue="quality"
)
plt.show()
```

```python
sns.barplot(
    data=df,
    x="quality",
    y="alcohol",
    estimator="mean"
)
plt.title("Average Alcohol by Wine Quality")
plt.show()
```

```python
g = sns.FacetGrid(df, col="quality", col_wrap=4)
g.map(sns.scatterplot, "alcohol", "pH")
plt.show()
```
